```python
from dnafiber.postprocess.error_detection import load_model
```

```python
model = load_model()
import torch.nn as nn
import timm
from huggingface_hub import PyTorchModelHubMixin


class MyModel(nn.Module, PyTorchModelHubMixin):
    def __init__(self):
        super().__init__()
        
        self.model = timm.create_model("seresnet50.a1_in1k",
                                       pretrained=False,
            num_classes=1,
            in_chans=4,
        )

    def forward(self, x):
        return self.model(x)

model_hub = MyModel()
model_hub.model.load_state_dict(model.state_dict())
model_hub.push_to_hub("ClementP/dnafiber-error-detection")
```
